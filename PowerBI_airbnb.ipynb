# Extracting basic details from Mongodb sample_airbnb Database

```python
_id=[]
listing_url=[]
name=[]
summary=[]
space=[]
description=[]
neighborhood_overview=[]
notes=[]
transit=[]
access=[]
interaction=[]
house_rules=[]
property_type=[]
room_type=[]
bed_type=[]
minimum_nights=[]
maximum_nights=[]
cancellation_policy=[]
last_scraped=[]
calendar_last_scraped=[]
first_review=[]
last_review=[]
accommodates=[]
bedrooms=[]
beds=[]
number_of_reviews=[]
bathrooms=[]
amenities=[]
price=[]
security_deposit=[]
cleaning_fee=[]
extra_people=[]
guests_included=[]

for x1 in records.find():

    _id.append(x1['_id'])
    listing_url.append(x1['listing_url'])
    name.append(x1['name'])
    summary.append(x1['summary'])
    space.append(x1['space'])
    description.append(x1['description'])
    neighborhood_overview.append(x1['neighborhood_overview'])
    notes.append(x1['notes'])
    transit.append(x1['transit'])
    access.append(x1['access'])
    interaction.append(x1['interaction'])
    house_rules.append(x1['house_rules'])
    property_type.append(x1['property_type'])
    room_type.append(x1['room_type'])
    bed_type.append(x1['bed_type'])
    minimum_nights.append(x1['minimum_nights'])
    maximum_nights.append(x1['maximum_nights'])
    cancellation_policy.append(x1['cancellation_policy'])
    last_scraped.append(x1['last_scraped'])
    calendar_last_scraped.append(x1['calendar_last_scraped'])
    try:
      first_review.append(x1['first_review'])
    except KeyError:
      first_review.append(np.nan)
    try:
      last_review.append(x1['last_review'])
    except KeyError:
      last_review.append(np.nan)
    try:
      accommodates.append(x1['accommodates'])
    except KeyError:
      accommodates.append(np.nan)
    try:
      bedrooms.append(x1['bedrooms'])
    except KeyError:
      bedrooms.append(np.nan)
    try:
      beds.append(x1['beds'])
    except KeyError:
      beds.append(np.nan)
    number_of_reviews.append(x1['number_of_reviews'])
    try:
      bathrooms.append(x1['bathrooms'])
    except KeyError:
      bathrooms.append(np.nan)
    try:
      amenities.append(x1['amenities'])
    except KeyError:
      amenities.append(np.nan)
    price.append(x1['price'])
    try:
      security_deposit.append(x1['security_deposit'])
    except KeyError:
      security_deposit.append(np.nan)
    try:
      cleaning_fee.append(x1['cleaning_fee'])
    except KeyError:
      cleaning_fee.append(np.nan)
    try:
      extra_people.append(x1['extra_people'])
    except KeyError:
      extra_people.append(np.nan)
    try:
      guests_included.append(x1['guests_included'])
    except KeyError:
      guests_included.append(np.nan)



master_df=pd.DataFrame({'_id':_id,
                        'listing_url':listing_url,
                        'name':name,
                        'summary':summary,
                        'space':space,
                        'description':description,
                        'neighborhood_overview':neighborhood_overview,
                        'notes':notes,
                        'transit':transit,
                        'access':access,
                        'interaction':interaction,
                        'house_rules':house_rules,
                        'property_type':property_type,
                        'room_type':room_type,
                        'bed_type':bed_type,
                        'minimum_nights':minimum_nights,
                        'maximum_nights':maximum_nights,
                        'cancellation_policy':cancellation_policy,
                        'last_scraped':last_scraped,
                        'calendar_last_scraped':calendar_last_scraped,
                        'first_review':first_review,
                        'last_review':last_review,
                        'accommodates':accommodates,
                        'bedrooms':bedrooms,
                        'beds':beds,
                        'number_of_reviews':number_of_reviews,
                        'bathrooms':bathrooms,
                        'amenities':amenities,
                        'price':price,
                        'security_deposit':security_deposit,
                        'cleaning_fee':cleaning_fee,
                        'extra_people':extra_people,
                        'guests_included':guests_included})

master_df.to_csv('master.csv', encoding='utf-8', index=False)
```

# Extracting address details from Mongodb sample_airbnb Database

```python
_id=[]
street=[]
suburb=[]
government_area=[]
market=[]
country=[]
country_code=[]
type_=[]
coordinates_lat=[]
coordinates_lon=[]
is_location_exact=[]

for x1 in records.find():
    a1=(list(x1.keys()))[-4]

    _id.append(x1['_id'])
    try:
      street.append(x1[a1]['street'])
    except KeyError:
      street.append(None)
    try:
      suburb.append(x1[a1]['suburb'])
    except KeyError:
      suburb.append(None)
    try:
      government_area.append(x1[a1]['government_area'])
    except KeyError:
      government_area.append(None)
    try:
      market.append(x1[a1]['market'])
    except KeyError:
      market.append(None)
    try:
      country.append(x1[a1]['country'])
    except KeyError:
      country.append(None)
    try:
      country_code.append(x1[a1]['country_code'])
    except KeyError:
      country_code.append(None)
    try:
      type_.append(x1[a1]['location']['type'])
    except KeyError:
      type_.append(None)
    try:
      coordinates_lat.append(x1[a1]['location']['coordinates'][0])
    except KeyError:
      coordinates_lat.append(None)
    try:
      coordinates_lon.append(x1[a1]['location']['coordinates'][1])
    except KeyError:
      coordinates_lon.append(None)
    try:
      is_location_exact.append(x1[a1]['location']['is_location_exact'])
    except KeyError:
      is_location_exact.append(None)

address_df=pd.DataFrame({'_id': _id,
                      'street':street,
                      'suburb':suburb,
                      'government_area':government_area,
                      'market':market,
                      'country':country,
                      'country_code':country_code})

address_df.to_csv('address.csv', encoding='utf-8', index=False)

location_df=pd.DataFrame({'_id':_id,
                          'type':type_,
                          'coordinates_lat':coordinates_lat,
                          'coordinates_lon':coordinates_lon,
                          'is_location_exact':is_location_exact})
location_df.to_csv('location.csv', encoding='utf-8', index=False)


```

## Extracting review details from Mongodb sample_airbnb Database

```python
_id=[]
date=[]
listing_id=[]
reviewer_id=[]
reviewer_name=[]
comments=[]

for x1 in records.find():
    a1=(list(x1.keys()))[-1]
    b1=len(x1[a1])
    for k in range(b1):
      _id.append(x1[a1][k]['_id'])
      try:
        date.append(x1[a1][k]['date'])
      except KeyError:
        date.append(None)
      try:
        listing_id.append(x1[a1][k]['listing_id'])
      except KeyError:
        listing_id.append(None)
      try:
        reviewer_id.append(x1[a1][k]['reviewer_id'])
      except KeyError:
        reviewer_id.append(None)
      try:
        reviewer_name.append(x1[a1][k]['reviewer_name'])
      except KeyError:
        reviewer_name.append(None)
      try:
        comments.append(x1[a1][k]['comments'])
      except KeyError:
        comments.append(None)


reviews_df=pd.DataFrame({'_id': _id,
                      'date':date,
                      'listing_id':listing_id,
                      'reviewer_id':reviewer_id,
                      'reviewer_name':reviewer_name,
                      'comments':comments})

reviews_df.to_csv('reviews.csv', encoding='utf-8', index=False)
```
